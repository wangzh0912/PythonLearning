```python
from bs4 import BeautifulSoup
import requests
import pandas as pd


factor_list = [
    'EUEPUINDXM',
    'EXUSUK',
    'GBRCPIALLMINMEI',
    'GBRCPICORMINMEI',
    'GBRPROINDMISMEI',
    'GBRRECM',
    'GBRSLRTCR03IXOBM',
    'IR3TIB01GBM156N',
    'IRLTLT01GBM156N',
    'LMJVTTUVGBM647S',
    'LRHUTTTTGBM156S',
    'MABMM301GBM189S',
    'MANMM101GBM189S',
    'RBGBBIS',
    'SPASTT01GBM661N',
    'LFWA64TTGBQ647S',	
    'QGBN628BIS',
    'QGBR628BIS',
    'UKNGDP',
]

enriched_factor = []
for factor in factor_list:
    url = 'https://fred.stlouisfed.org/series/' + factor
    soup = BeautifulSoup(requests.get(url).content, 'html.parser')
    unit = soup.body.find('span', attrs={'class': 'series-meta-value-units'})
    freq = soup.body.find('span', attrs={'class': 'series-meta-value-frequency'})
    name = soup.body.find('span', attrs={'id': 'series-title-text-container'})
    name = name.text.split('(')[0].strip()
    res = {'symbol': factor, 'name': name, 'freq': freq.text.strip(), 'raw_data_type': unit.text.strip()}
    enriched_factor.append(res)

enriched_factor = pd.DataFrame(enriched_factor)
```
